# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
total_ride_count_city_type = pyber_data_df.groupby(["type"]).count()["ride_id"]

total_ride_count_city_type.head()
```

```python
# 2. Get the total drivers for each city type
total_driver_sum_city_type = city_data_df.groupby(["type"]).sum()["driver_count"]

total_driver_sum_city_type
```

```python
#  3. Get the total amount of fares for each city type
total_fare_sum_city_type = pyber_data_df.groupby(["type"]).sum()["fare"]

total_fare_sum_city_type 
```

```python
#  4. Get the average fare per ride for each city type.
average_fare_per_ride_city_type = total_fare_sum_city_type/total_ride_count_city_type

average_fare_per_ride_city_type

```

```python
# 5. Get the average fare per driver for each city type. 
average_fare_per_driver_city_type = total_fare_sum_city_type/total_driver_sum_city_type

average_fare_per_driver_city_type



```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df=pd.DataFrame()

pyber_summary_df["Total Rides"] = total_ride_count_city_type
pyber_summary_df["Total Drivers"] = total_driver_sum_city_type
pyber_summary_df["Total Fares"] = total_fare_sum_city_type 
pyber_summary_df["Average Fare per Ride"] = average_fare_per_ride_city_type
pyber_summary_df["Average Fare per Driver"] = average_fare_per_driver_city_type


pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None

pyber_summary_df
```

```python
#  8. Format the columns.

pyber_summary_df["Total Rides"] = total_ride_count_city_type.map("{:,}".format)
pyber_summary_df["Total Drivers"] = total_driver_sum_city_type.map("{:,}".format)
pyber_summary_df["Total Fares"] = total_fare_sum_city_type.map("${:,.2f}".format)
pyber_summary_df["Average Fare per Ride"] = average_fare_per_ride_city_type.map("${:,.2f}".format)
pyber_summary_df["Average Fare per Driver"] = average_fare_per_driver_city_type.map("${:,.2f}".format)

pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
sum_fares_date_type = pyber_data_df.groupby(["type","date"]).sum()[["fare"]]

sum_fares_date_type
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()

sum_fares_date_type = sum_fares_date_type.reset_index()

sum_fares_date_type
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 

sum_fares_date_type_pivot = sum_fares_date_type.pivot(index = "date", columns = "type", values = "fare")


sum_fares_date_type_pivot 
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
jan_apr_fares_df=sum_fares_date_type_pivot.loc['2019-01-01':'2019-04-29']

jan_apr_fares_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)

jan_apr_fares_df.index = pd.to_datetime(jan_apr_fares_df.index)

jan_apr_fares_df
```

```python
# 7. Check that the datatype for the index is datetime using df.info()

jan_apr_fares_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.

weekly_fares_df = jan_apr_fares_df.resample("W").sum()

weekly_fares_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

#Plot
weekly_fares_df.plot()

#Edit the Plot

plt.title("Total Fare by City Type")
plt.ylabel("Fare ($USD)")
plt.xlabel(" ")

#Save the Plot
plt.savefig("Analysis/PyBer_fare_summary.png")
```
